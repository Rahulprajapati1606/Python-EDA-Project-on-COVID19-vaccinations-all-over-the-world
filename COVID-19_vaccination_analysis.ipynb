## Exploratory Data Analysis

### Total Vaccinations

```python
# plotting a bar chart for Country vs total Vaccinations
TV_Country = df.groupby(['country'],as_index=False)['total_vaccinations'].max().sort_values(by= 'total_vaccinations',ascending=False).head(10)
sns.set({'figure.figsize':(15,5)})
ax = sns.barplot(data=TV_Country,x='country',y='total_vaccinations')
for bars in ax.containers:
    ax.bar_label(bars,labels=[f'{int(bar.get_height())/1e9:.2f}B' for bar in bars])
```

##### From above graph it is clear that China has the maximum number of total vaccinations over the time followed by India and       US

## People Vaccinated and People Fully Vaccinated

```python
PV_Country = df.groupby('country',as_index=False)[['people_vaccinated','people_fully_vaccinated']].max().sort_values(by='people_vaccinated', ascending=False).head(10)
PV_Country = PV_Country.melt(id_vars='country',value_vars=['people_vaccinated','people_fully_vaccinated'],var_name='status',value_name='count')
sns.lineplot(data = PV_Country,x = 'country',y =  'count',hue='status',marker='o')
```

##### From above line plot a detailed comparision between people vaccinated and fully vaccinated according to the country can        be seen easily and data shows that China has the maximum no. of people being vaccinated and fully vaccinated followed by       India and US.

## Daily Vaccinations and the type of vaccines used in each country

```python
DV_Country = df.groupby(['country','vaccines'],as_index=False)['daily_vaccinations'].sum().sort_values(by= 'daily_vaccinations',ascending=False).head(10)
ax = sns.barplot(data=DV_Country,x='country',y='daily_vaccinations',hue='vaccines')
for bars in ax.containers:
    ax.bar_label(bars,labels=[f'{int(bar.get_height())/1e9:.2f}B' for bar in bars])
```

```python
top_vaccines = df['vaccines'].value_counts().head(5).index
filter_df = df[df['vaccines'].isin(top_vaccines)] 
sns.set(rc={'figure.figsize':(28,5)})
ax = sns.countplot(x='vaccines',hue='vaccines',data = filter_df)
for bars in ax.containers:
    ax.bar_label(bars)
```

##### From above graphs it can be seen that in terms of daily vaccinations India is on the top followed by US and Bazil and as       per the data Johnson&Johnson, Oxford/AstraZeneca, Pfizer/BioNTech, and Moderna are the most used vaccines but their            individual use is not mentioned in the data so it can't be commented.


## Total_vaccinations_per_hundred,
## People_vaccinated_per_hundred,
## People_fully_vaccinated_per_hundred

```python
PH_Country = df.groupby('country',as_index=False)[['total_vaccinations_per_hundred','people_vaccinated_per_hundred','people_fully_vaccinated_per_hundred']].max().sort_values(by = 'total_vaccinations_per_hundred',ascending= False).head(10)
PH_Country = PH_Country.melt(id_vars='country',value_vars=['total_vaccinations_per_hundred','people_vaccinated_per_hundred','people_fully_vaccinated_per_hundred'],var_name='status',value_name='count')
sns.lineplot(data = PH_Country,x='country',y='count',hue='status',marker='o')  
sns.set(rc={'figure.figsize':(10,3)})
```

##### From above graph it is concluded that Gibraltar had the maximum number of people_vaccinated_per_hundred and maximum            people_fully_vaccinated_per_hundred and also maximum total_vaccinations_per_hundred

## Daily_vaccinations_per_million

```python
PM_Country = df.groupby('country',as_index=False)['daily_vaccinations_per_million'].sum().sort_values('daily_vaccinations_per_million',ascending=False).head(10)
sns.lineplot(data = PM_Country,x='country',y='daily_vaccinations_per_million',marker='o')
```

##### So as per the graph obtained Cuba has the maximum Daily_vaccinations_per_million during the pandemic time
